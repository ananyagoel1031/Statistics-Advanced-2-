### Q1. What is hypothesis testing in statistics?

#### A1. Hypothesis testing is a statistical method used to make inferences about population parameters based on sample data, allowing researchers to determine the validity of a hypothesis.

#### Hypothesis testing is a formal procedure in statistics that evaluates two mutually exclusive statements about a population: the null hypothesis (H₀) and the alternative hypothesis (H₁). The null hypothesis typically states that there is no effect or no difference, while the alternative hypothesis suggests that there is an effect or a difference. The primary purpose of hypothesis testing is to determine whether there is enough evidence in the sample data to reject the null hypothesis in favor of the alternative hypothesis.

### Q4. What are Type I and Type II errors? Give examples of each. 

#### A4. In hypothesis testing, Type I and Type II errors are critical concepts that describe the potential mistakes made when interpreting statistical results. These errors revolve around the truth or falsity of the null hypothesis (H₀).

#### Type I Error: False Positive
#### A Type I error occurs when a true null hypothesis is incorrectly rejected. This means we conclude there is an effect or difference when, in reality, there is none. It is also referred to as a false positive.
#### For example:
#### In medical testing, diagnosing a healthy person as having a disease is a Type I error.
#### In a legal trial, convicting an innocent person is a Type I error.

#### Type II Error: False Negative
#### A Type II error happens when a false null hypothesis is not rejected. This means we fail to detect an effect or difference that actually exists. It is also called a false negative.
#### For example:
#### In medical testing, failing to diagnose a sick person is a Type II error.
#### In quality control, accepting a defective product as meeting standards is a Type II error.

### Q5. What is the difference between a Z-test and a T-test? Explain when to use each.

#### A5. A Z-test is used when the population variance is known and the sample size is large, while a T-test is used when the population variance is unknown and the sample size is small.

#### Z-test: A statistical test used to determine if there is a significant difference between sample and population means or between the means of two samples when the population variance is known and the sample size is large (typically n≥30). It relies on the standard normal distribution (Z-distribution). 

#### T-test: A statistical test used to determine if there is a significant difference between the means of two groups, particularly when the population variance is unknown and the sample size is small (typically n<30). It relies on the t-distribution, which has heavier tails than the normal distribution. 


### Q6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram. (Include your Python code and output in the code box below.) 

```python
#A6. 

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success
size = 1000  # Number of experiments

# Generate binomial distribution data
data = np.random.binomial(n, p, size)

# Plot the histogram
plt.hist(data, bins=range(n+2), density=True, alpha=0.75, color='blue', edgecolor='black')
plt.title('Histogram of Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

#### Output:
#### The histogram will display the probability distribution of the number of successes in 10 trials, with a peak around 5 (the expected value for ( n=10, p=0.5 )).

### Q7. : Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9] (Include your Python code and output in the code box below.)

```python
#A7.
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 
               49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3,
               50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0,
               49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
n = len(sample_data)

# Calculate Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Z-Statistic: {z_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Decision at significance level alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

### Q8.  Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. (Include your Python code and output in the code box below.)

```python
#A8.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
mean = 50  # Mean of the distribution
std_dev = 10  # Standard deviation of the distribution
sample_size = 1000  # Number of data points

data = np.random.normal(mean, std_dev, sample_size)

# Step 2: Calculate the 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
margin_of_error = 1.96 * (sample_std / np.sqrt(sample_size))  # 95% confidence level

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Step 3: Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7, density=True)

# Overlay the normal distribution curve
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
plt.plot(x, norm.pdf(x, mean, std_dev), color='red', label='True Distribution')

plt.title('Histogram of Simulated Data with Normal Distribution Curve')
plt.xlabel('Value')
plt.ylabel('Density')
plt.axvline(confidence_interval[0], color='green', linestyle='--', label='95% CI Lower Bound')
plt.axvline(confidence_interval[1], color='green', linestyle='--', label='95% CI Upper Bound')
plt.legend()
plt.show()

# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: {confidence_interval}")
```

### Q9.  Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean. (Include your Python code and output in the code box below.) 

```python
#A9.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_visualize_zscores(data):
    """
    This function calculates Z-scores for a dataset and visualizes the standardized data using a histogram.

    Parameters:
        data (list or numpy array): The input dataset.

    Returns:
        z_scores (numpy array): The calculated Z-scores.
    """
    # Calculate Z-scores
    z_scores = zscore(data)
    
    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores', fontsize=14)
    plt.xlabel('Z-score', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1, label='Mean (Z=0)')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()
    
    return z_scores

# Example usage
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]
z_scores = calculate_and_visualize_zscores(data)

# Print Z-scores
print("Z-scores:", z_scores)
```
